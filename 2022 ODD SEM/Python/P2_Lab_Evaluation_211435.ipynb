# Question 1:

```python
'''L is a list defined as L= {&quot;One&quot;,&quot;Two&quot;,&quot;Three&quot;,&quot;Four&quot;,&quot;Five&quot;}. WAP to count the length
of reach element from a list and write to the file in the given below format:
One, 3
Two, 3
Four, 4'''

L = ["One", "Two", "Three", "Four", "Five"]
with open("ques1.txt", "w+") as infile:
  for i in L:
    a = len(i)
    b = i + ", " + str(a) + "\n"
    infile.write(b)
    #just to check what is being printed
    print(b, end = "")
```

# Question 2:

```python
'''Write a Python program to create a file where all letters of the English alphabet are listed
by specified number of letters on each line.'''

import string
def ques2(n):
   with open("ques2.txt", "w") as f:
       alpha = string.ascii_uppercase
       letters = [alpha[i:i + n] + "\n" for i in range(0, len(alpha), n)]
       f.writelines(letters)
       #just to check what is being printed
       print(letters)

a = int(input("Ënter the number of alphabets to be written in one line : "))
ques2(a)
```

```python
'''Write a Python program to read a file line by line and store it into a list.'''

with open("ques2.txt", "r") as infile:
  a = list()
  for line in infile:
    a.append(line)
  print(a)
```

# Question 4:

```python
'''Write to the file 100 random strings whose length is between 10 and 15.'''

import random

with open("ques4.txt", "w+") as infile:
  for i in range(1, 101):
    rng = [10, 11, 12, 13, 14, 15]
    N = random.choice(rng)
    currstr = ''.join(random.choices(string.ascii_uppercase + string.ascii_lowercase, k=N))
    currstr = currstr + "\n"
    infile.write(currstr)
    #just to check what is being printed
    print(currstr, end = "")
```
