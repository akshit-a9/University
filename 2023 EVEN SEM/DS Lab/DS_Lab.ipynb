## Simpy

2 - To create a car process in simpy where car is parking and driving /
Simply getting started with simpy using car process where car is being driven and parked.

```python
def car(env):
  while True:
    print('parking of car starts at %d' %env.now)
    parking_duration=4
    yield env.timeout(parking_duration)

    print('driving of car starts at %d' %env.now)
    trip_duration=3
    yield env.timeout(trip_duration)
```

```python
import simpy
env=simpy.Environment()
env.process(car(env))
env.run(until=20)
```

3 - To make a program in simpy for interaction where waiting method is utilised

```python
class Car:
  def __init__(self, env):
    self.env = env
    self.action = env.process(self.drive())

  def drive(self):
    while True:
      print("The Car is at parking and charging at the time: %d" % self.env.now)
      parking_duration = 7
      yield self.env.process(self.charge(parking_duration))

      print('The car is driving at %d' %self.env.now)
      drive_duration = 3
      yield self.env.timeout(drive_duration)

  def charge(self, duration):
    yield self.env.timeout(duration)
```

```python
import simpy
env = simpy.Environment()
Car(env)
env.run(until=100)
```

# **20/01/2023**

```python
import simpy
def driver(env, car):
  yield env.timeout(3)
  car.action.interrupt()
```

```python
class Car(object):
  def __init__(self, env):
    self.env = env
    self.action = env.process(self.run())

  def run(self):
    while True:
      print('Start parking and charging at %d' % self.env.now)
      charge_duration = 7

      try:
        yield self.env.process(self.charge(charge_duration))
      except simpy.Interrupt:
        print('Was interrupted. Hope, the battery is full enough ...')

      print('Start driving at %d' % self.env.now)
      trip_duration = 3
      yield self.env.timeout(trip_duration)

  def charge(self, duration):
    yield self.env.timeout(duration)
```

```python
env = simpy.Environment()
car = Car(env)
env.process(driver(env, car))
env.run(until=85)
```

Shared Resources in Simpy

```python
def car(env, name, bcs, driving_time, charge_duration):
  yield env.timeout(driving_time)
  print('%s arriving at %d' % (name, env.now))
  with bcs.request() as req:
    yield req
    print('%s starting to charge at %s' %(name, env.now))
    yield env.timeout(charge_duration)
    print('%s leaving the bcs at %s' % (name, env.now))
```

```python
import simpy
env = simpy.Environment()
bcs = simpy.Resource(env, capacity = 4)
```

```python
for i in range(7):
  env.process(car(env, 'car %d' % i, bcs, i*2, 5))
```

```python
env.run()
```

```python
def resource_user(name, env, resource, wait, prio):
  yield env.timeout(wait)
  with resource.request(priority = prio) as req:
    print(f'p{name} requesting at {env.now} with priority = {prio}')
    yield req
    print(f'p{name} got resource at {env.now}')
    yield env.timeout(6)
```

Implementing of preemptive resource in simpy

```python
import simpy
def resource_user(name, env, resource, wait, prio):
  yield env.timeout(wait)
  with resource.request(priority = prio) as req:
    print(f'{name} requesting at {env.now} with priority = {prio}')
    yield req
    print(f'{name} got resource at {env.now}')

    try:
      yield env.timeout(2)
    except simpy.Interrupt as interrupt:
      by = interrupt.cause.by
      usage = env.now - interrupt.cause.usage_since
      print(f'{name} got preempted by {by} at {env.now}'f' after {usage}')
```

```python
env = simpy.Environment()
res = simpy.PreemptiveResource(env, capacity = 1)
p1 = env.process(resource_user(1, env, res, wait = 10, prio = 0))
p2 = env.process(resource_user(2, env, res, wait = 9, prio = 0))
p3 = env.process(resource_user(3, env, res, wait = 8, prio = 1))
p4 = env.process(resource_user(4, env, res, wait = 1, prio = 3))
env.run()
```

```python
def user(name, env, res, prio, preempt):
  with res.request(priority = prio, preempt = preempt) as req:
    try:
      print(f'{name} requesting at {env.now}')
      assert isinstance(env.now, int), type(env.now)
      yield req
      assert isinstance(env.now, int), type(env.now)
      print(f'{name} got resource at {env.now}')
      yield env.timeout(3)
    except simpy.Interrupt:
      print(f'{name} got preempted at {env.now}')
```

```python
env = simpy.Environment()
res = simpy.PreemptiveResource(env, capacity = 1)
A = env.process(user('A', env, res, prio = 0, preempt=True))
B = env.process(user('B', env, res, prio = -2, preempt=False))
C = env.process(user('C', env, res, prio = 1, preempt=True))
D = env.process(user('D', env, res, prio = 3, preempt=False))
env.run()
```

To simulate a real wall clock in simpy

```python
import time
from datetime import datetime
```

```python
def timeexp(env):
  start = env.now
  print('The process has started at %s' %datetime.now())
  yield env.timeout(7)
  print('The process has finished at %s' %datetime.now())
  end = env.now
  print('The duration of the process was %d' %(end-start))
```

```python
import simpy
env = simpy.Environment()
env.process(timeexp(env))
env.run()
```

```python
def timeexp(env):
  start = time.perf_counter()
  print('The process has started at %s' %datetime.now())
  yield env.timeout(7)
  print('The process has finished at %s' %datetime.now())
  end = time.perf_counter()
  print('The duration of the process in wall clock was %d seconds' %(end-start))
```

```python
import simpy
env = simpy.Environment()
env.process(timeexp(env))
env.run()
```

```python
import simpy.rt as rtt
```

```python
env = rtt.RealtimeEnvironment(initial_time = 0, factor = 1, strict = False)
env.process(timeexp(env))
env.run()
```

```python
def slow_process(env):
  time.sleep(15)   #14
  yield env.timeout(2)
```

```python
env = rtt.RealtimeEnvironment(factor = 5, strict = True) #also check with false
env.process(slow_process(env))
try:
  env.run()
  print('Everything remained smooth')
except RuntimeError:
  print('The process was too slow compared to the wall clock')
#if the difference is less than factor, it will ignore the difference and remain smooth
```
